```r
# Read the data into a data frame
# "<-" is used as an assignment operator:
df <- read.csv("https://raw.githubusercontent.com/cjflanagan/sample_data/master/grad-students.csv")
```

```r
# head function prints first few rows and all the columns
head(df)
```

```r
# Similarly the tail function prints last few rows and all the columns
tail(df)
```

```r
# dim function will print the dimensions of the data frame
# 173 rows and 22 columns
dim(df)
```

```r
# "=" can also be used as an assignment operator:
# We can also use "read.delim" function to read the file
# We must specify the separator when using "read.delim" function


stem_women = read.delim("https://raw.githubusercontent.com/cjflanagan/sample_data/master/women-stem.csv", sep=",")
```

```r
# Let's find the contents of this d
head(stem_women)
```

```r
# Print the dimensions of the data frame
dim(stem_women)
```

```r
# Print all the column names in the data frame
colnames(stem_women)
```

```r
# Suppose we want to only get the columns Major, Major_category, Men and Women
# Suppose we want to get only 20 rows of these
stem_women[1:20, c(3,4,6,7)]
```

```r
# Suppose we want to only get the columns Major, Major_category, Men and Women
# Suppose we want to get only 20 rows of these
# Suppose we want to use the column names to subset and not the column numbers
stem_women[1:20, c("Major", "Major_category", "Men", "Women")]
```

```r
# Note c(3,4,6,7) → the “c” stands for concatenate function. It doesn’t indicate columns. 

stem_women[c(2,3,4,7,8,9,11,24), c(3,4,6,7)]

# prints rows 2,3,4,7,8,9,11 and 24 and columns 3,4,6 and 7
```

```r
#ggplot2 is the most commonly used library for data visualization in R

library(ggplot2)
```

```r
ggplot(stem_women, aes(x=Men, y=Women, colour = Major_category)) + 
  geom_point()
```

```r
ggplot(stem_women, aes(x=log(Men), y=log(Women), colour = Major_category)) + 
  geom_point() + theme_minimal()
```

```r
ggplot(stem_women, aes(x=Major_category, y=log(Total), color=Major_category, fill=Major_category)) +
  geom_violin() +theme_minimal()
```

```r
ggplot(stem_women, aes(x=Major_category, y=Total, color=Major_category, fill=Major_category)) +
  geom_violin() +theme_minimal() +coord_flip()
```
