```r
# Read the data file
# "<-" is used as an assignment operator:
myfile <- read.csv("https://raw.githubusercontent.com/meenakshikagda/Data-visualization/master/datasets/kaggle-breast-cancer-dataset/breast_cancer_wisconsin_diagnostic_dataset_kaggle_download.csv")

## From Kaggle Breast cancer dataset
```

```r
# Suppose we want to only get the columns id, diagnosis, radius_mean, perimenter_mean
# Suppose we want to just see the top 20 rows of these columns
# In R, index starts from 1 unlike 0 in python
# Always first parameter passed will refer rows and the second will refer to columns

myfile[1:20, c(2,3,5)]
```

```r
# Lets save this in a new data frame:

df <- myfile[sample(1:569, 50), c(2,3,5)]
```

```r
# basic ggplot graph
ggplot(df, aes(x=radius_mean, y=perimeter_mean, color=diagnosis)) + geom_point()
```

```r
# lets save it in a object p

p <- ggplot(df, aes(x=radius_mean, y=perimeter_mean, color=diagnosis)) + geom_point()
```

```r
# lets get rid of the background grey color

p + theme_bw()
```

```r
## Using ggthemes, we can use economist theme

p + theme_economist()
```

```r
# Lets change it to economist white theme

p + theme_economist_white()
```

```r
p + theme_economist_white() + ggtitle("Radius mean vs perimeter mean of cells in breast cancer")
```

```r
# lets plot the entire dataset

p + geom_point(size=3) + theme_economist_white()
```

```r
# Add linear regression line 
#  (by default includes 95% confidence region)

ggplot(df, aes(x=radius_mean, y=perimeter_mean, color=diagnosis)) + 
    geom_point() + geom_smooth(method='lm') +
    theme_economist_white()

```

```r
# Lets redo this on entire data
# Add linear regression line 
#  (by default includes 95% confidence region)
# lets add fivethirtyeight colors

ggplot(myfile, aes(x=radius_mean, 
                    y=perimeter_mean, 
                    color=diagnosis)) + 
    geom_point() + 
    geom_smooth(method='lm', color="black") +
    theme_economist_white() +
    scale_colour_fivethirtyeight()
```

```r
# Let's try to see if we can change the size of the points
# Lets try perimenter_mean
# let's change alpha (transperancy of the points)


ggplot(myfile, aes(x=radius_mean, 
                    y=perimeter_mean, 
                    color=diagnosis, 
                    size=perimeter_mean, alpha=0.8)) + 
    geom_point() + 
    #geom_smooth(method='lm', color="black") +
    theme_fivethirtyeight() +
    scale_colour_fivethirtyeight()
```

```r
# Lets see how it looks withouth the alpha
# lets change regression line into a dashed line

ggplot(myfile, aes(x=radius_mean, 
                    y=perimeter_mean, 
                    color=diagnosis, 
                    size=radius_mean)) + 
    geom_point() + 
    geom_smooth(method='lm', color="black", linetype="dashed") +
    theme_fivethirtyeight() +
    scale_colour_fivethirtyeight()
```

```r
# After say all the combinations, I just like the basic size of the dots
# So lets save it as plot in a variable called plot

plot <- ggplot(myfile, aes(x=radius_mean, 
                    y=perimeter_mean, 
                    color=diagnosis)) + 
    geom_point() + 
    geom_smooth(method='lm', color="black", linetype="dashed") +
    theme_economist_white() +
    scale_colour_fivethirtyeight()
```

```r
# To see the plot:
plot
```

```r

# Lets move the legends to the bottom

plot+theme(legend.position="bottom")
```

```r
# We can also move the legends to the right

plot+theme(legend.position="right")
```

```r
# we can also move it inside the plot
plot+ theme(legend.position = c(0.8, 0.2))
```

```r
# we can also move it inside the plot
plot+ theme(legend.position = c(0.7,0.9))
```

```r
plot+ theme(legend.position = "bottom") +
    ggtitle("Radius vs Perimeter of cancer cells by diagnosis type")
```

```r
# Let's play with colors
##https://www.nceas.ucsb.edu/~frazier/RSpatialGuides/colorPaletteCheatsheet.pdf


mynamestheme <- theme(plot.title = element_text(family = "Helvetica", face = "bold", size = (15), color ="navyblue"), 
                 legend.title = element_text(colour = "Seagreen",  face = "bold.italic", family = "Helvetica"), 
                 legend.text = element_text(face = "italic", colour="Red4",family = "Helvetica"), 
                  axis.title = element_text(family = "Helvetica", size = (15), colour = "skyblue", face="bold"),
                  axis.text = element_text(family = "Courier", colour = "violetred", size = (10)))
  
```

```r
# lets add it to the plot
plot + mynamestheme +
    theme(legend.position = "bottom") +
    ggtitle("Radius vs Perimeter of cancer cells by diagnosis type") +
    theme(legend.background = element_blank(),
        legend.box.background = element_rect(colour = "black"))
```

```r
# Let's look at some bar plots:
# We will look at some random 25 rows
# subset the entire dataset into another dataset

myfile2 <- myfile[sample(1:569, 25), ]
head(myfile2)
```

```r
colnames(myfile2)
```

```r
# Lets plot these into a horizontal bar plot
# as.character will ensure that the ids are treated as string and not numbers
# This dataset also contains Standard Error information.
# So we can use that in our plot

ggplot(myfile2) +
    geom_bar(aes(x=as.character(id), 
                    y=radius_mean, 
                    color=diagnosis, 
                    fill=diagnosis), 
                    stat="identity", alpha=0.7) +
    geom_errorbar( aes(x=as.character(id), 
                    ymin=radius_mean-radius_se, 
                    ymax=radius_mean+radius_se), 
                    width=0.3, colour="black", alpha=0.9, size=1.3) +
    theme_economist_white()
```

```r
# Let's flip x-axis and y-axis using coord_flip()
ggplot(myfile2) +
    geom_bar(aes(x=as.character(id), 
                    y=radius_mean, 
                    color=diagnosis, 
                    fill=diagnosis), 
                    stat="identity", alpha=0.7) +
    geom_errorbar( aes(x=as.character(id), 
                    ymin=radius_mean-radius_se, 
                    ymax=radius_mean+radius_se), 
                    width=0.3, colour="black", alpha=0.9, size=1.3) +
    theme_economist_white() + coord_flip()

    #coord_polar()
```

```r
# Let's use coord_polar to see the round plot

ggplot(myfile2) +
    geom_bar(aes(x=as.character(id), 
                    y=radius_mean, 
                    color=diagnosis, 
                    fill=diagnosis), 
                    stat="identity", alpha=0.7) +
    geom_errorbar( aes(x=as.character(id), 
                    ymin=radius_mean-radius_se, 
                    ymax=radius_mean+radius_se), 
                    width=0.3, colour="black", alpha=0.9, size=1.3) +
    theme_bw() + coord_polar()
```

```r
# Since we added major grid lines, the x-axis text looks weired
# So let's remove x-axis text (just to see how it looks)
# Not the best plot but something to play around

plot2 + theme(panel.grid.major = element_line(colour = "black"),
        axis.text.x=element_blank(),
        axis.title.x=element_blank())
```

```r
# Let's look at histogram
ggplot(myfile, aes(x=radius_mean, color=diagnosis, fill=diagnosis)) + 
  geom_histogram(bins=30, alpha=0.8) +
  theme_bw() +
  ggtitle("Frequency distribution of radius") +
  theme(plot.title = element_text(hjust = 0.5)) +
  xlab("Radius mean") +
  ylab("Frequency")
```

```r
# Interleaved histograms
ggplot(myfile, aes(x=radius_mean, color=diagnosis, fill=diagnosis)) + 
            geom_histogram(bins=30, position="dodge", alpha=0.8) +
            theme_bw() +
            ggtitle("Frequency distribution of radius") +
            theme(plot.title = element_text(hjust = 0.5)) +
            xlab("Radius mean") +
            ylab("Frequency")
```

```r
# Interleaved histograms

ggplot(myfile, aes(x=perimeter_worst, color=diagnosis, fill=diagnosis)) + 
            geom_histogram(bins=30, position="dodge", alpha=0.8) +
            ggtitle("Frequency distribution of perimeter worst") +
            theme(plot.title = element_text(hjust = 0.5, vjust=0.5)) +
            xlab("Perimeter worst") +
            ylab("Frequency") +
            theme_economist_white() 
            
```
